```python
import numpy as np
from scipy.io import loadmat

# Load the SVHN training dataset
train_data = loadmat('train_32x32.mat')

# Extract the data and labels
X_train = train_data['X']
y_train = train_data['y']

# Reshape and normalize the data
X_train = np.transpose(X_train, (3, 0, 1, 2))  # Transpose to (num_samples, width, height, channels)
X_train = X_train.astype('float32') / 255.0  # Normalize pixel values

# Now you have the training data in 'X_train' and labels in 'y_train' in a suitable format.

# Continue with the code to randomly select samples from the training dataset as shown in the previous response.


# Now you have the training data in 'X_train' and labels in 'y_train'.

# Proceed with the code to randomly select samples from the training dataset as shown in the previous response.

# Assuming you have the full training data in 'X_train' and corresponding labels in 'y_train'
# 'X_train' should be a NumPy array where each row is a data point, and 'y_train' is a NumPy array of labels.

# Create an array to store the selected samples and labels
new_X_train = []
new_y_train = []

# Number of samples to select from each class
samples_per_class = 2500

# List of unique class labels (0 to 9)
unique_labels = np.unique(y_train)

# Set a random seed for reproducibility
np.random.seed(42)

# Randomly select samples from each class
for label in unique_labels:
    # Filter data points of the current class
    class_indices = np.where(y_train == label)[0]
    
    # Randomly choose 'samples_per_class' samples from this class
    selected_indices = np.random.choice(class_indices, samples_per_class, replace=False)
    
    # Append the selected samples and labels to the new training dataset
    new_X_train.extend(X_train[selected_indices])
    new_y_train.extend(y_train[selected_indices])

# Convert the lists to NumPy arrays
new_X_train = np.array(new_X_train)
new_y_train = np.array(new_y_train)

# Now, 'new_X_train' contains the 25,000 selected samples, and 'new_y_train' contains their corresponding labels.
```

```python
#Q1.a
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Assuming you have your randomly selected 25,000 training samples loaded into a variable 'new_X_train'

# Reshape the data to 2D
num_samples, width, height, channels = new_X_train.shape
new_X_train = new_X_train.reshape(num_samples, -1)

# Standardize the data (subtract mean and divide by standard deviation)
new_X_train = (new_X_train - np.mean(new_X_train, axis=0)) / np.std(new_X_train, axis=0)

# Initialize PCA
pca = PCA()

# Fit PCA to the selected training data
pca.fit(new_X_train)

# Calculate the cumulative explained variance
explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Find the number of top eigenvectors required to keep the proportion of variance above 0.9
num_eigenvectors = np.argmax(explained_variance_ratio >= 0.9) + 1

# Plot Proportion of Variance (PoV) against the number of eigenvectors
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.xlabel('Number of Eigenvectors')
plt.ylabel('Proportion of Variance')
plt.axvline(num_eigenvectors, color='red', linestyle='--', label=f'{num_eigenvectors} Eigenvectors (PoV >= 0.9)')
plt.legend()
plt.show()

print(f'Number of top eigenvectors required to keep PoV >= 0.9: {num_eigenvectors}')
```

```python
#Q2.a

# This code may take 30s to output the images, please wait.

import numpy as np
from scipy.io import loadmat
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the SVHN training dataset
train_data = loadmat('train_32x32.mat')

# Extract the data and labels
X_train = train_data['X']
y_train = train_data['y']

# Reshape and normalize the data
X_train = np.transpose(X_train, (3, 0, 1, 2))  # Transpose to (num_samples, width, height, channels)
X_train = X_train.astype('float32') / 255.0  # Normalize pixel values

# Reshape the data to 2D
num_samples, width, height, channels = X_train.shape
X_train_flat = X_train.reshape(num_samples, -1)

# Perform PCA on the raw data
pca = PCA()
X_train_pca = pca.fit_transform(X_train_flat)

# Find the number of top eigenvectors required to keep the proportion of variance above 0.9
num_eigenvectors = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.9) + 1

print(f'This code may take a while to output, please wait.')

# Visualize top 10 eigenvectors
top_eigenvectors = pca.components_[:10]
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
fig.suptitle('Top 10 Eigenvectors')

for i, ax in enumerate(axes.flat):
    eigenvector = top_eigenvectors[i].reshape(width, height, channels)
    normalized_eigenvector = (eigenvector - np.min(eigenvector)) / (np.max(eigenvector) - np.min(eigenvector))
    ax.imshow(normalized_eigenvector)
    ax.set_title(f'Eigenvector {i + 1}')
    ax.axis('off')

plt.show()

# Perform k-means clustering on raw data
kmeans_raw = KMeans(n_clusters=10, random_state=42, n_init=10)  # Set n_init explicitly
kmeans_raw.fit(X_train_flat)

# Perform k-means clustering on PCA-transformed data
kmeans_pca = KMeans(n_clusters=10, random_state=42, n_init=10)  # Set n_init explicitly
kmeans_pca.fit(X_train_pca[:, :num_eigenvectors])

# Function to get the index of the image closest to the centroid
def get_closest_image(centroid, data):
    distances = np.linalg.norm(data - centroid, axis=1)
    return np.argmin(distances)

# Display images closest to the centroids for raw data
for i, centroid in enumerate(kmeans_raw.cluster_centers_):
    closest_index = get_closest_image(centroid, X_train_flat)
    closest_image = X_train[closest_index]

    plt.subplot(2, 5, i + 1)
    plt.imshow(closest_image)
    plt.title(f'Cluster {i + 1}')
    plt.axis('off')

plt.suptitle('Images Closest to Centroids (Raw Data)')
plt.show()

# Display images closest to the centroids for PCA-transformed data
for i, centroid in enumerate(kmeans_pca.cluster_centers_):
    closest_index = get_closest_image(centroid, X_train_pca[:, :num_eigenvectors])
    closest_image = X_train[closest_index]

    plt.subplot(2, 5, i + 1)
    plt.imshow(closest_image)
    plt.title(f'Cluster {i + 1}')
    plt.axis('off')

plt.suptitle('Images Closest to Centroids (PCA-transformed Data)')
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics.cluster import contingency_matrix

# Function to compute purity
def purity_score(y_true, y_pred):
    contingency = contingency_matrix(y_true, y_pred)
    return np.sum(np.amax(contingency, axis=0)) / np.sum(contingency)

def compute_sse(data, centroids, labels):
    sse = 0.0
    for i in range(len(data)):
        centroid = centroids[labels[i]]
        sse += np.sum((data[i] - centroid) ** 2)
    return sse

# Ground truth labels (y_train) for evaluation
y_true = y_train.flatten() - 1  # Subtract 1 to make labels 0-based

# Evaluate clustering on raw data
sse_raw = compute_sse(X_train_flat, kmeans_raw.cluster_centers_, kmeans_raw.labels_)
purity_raw = purity_score(y_true, kmeans_raw.labels_)
rand_index_raw = adjusted_rand_score(y_true, kmeans_raw.labels_)

print(f'Sum Squared Error (Raw Data): {sse_raw}')
print(f'Purity (Raw Data): {purity_raw}')
print(f'Rand Index (Raw Data): {rand_index_raw}')

# Evaluate clustering on PCA-transformed data
sse_pca = compute_sse(X_train_flat, kmeans_pca.cluster_centers_, kmeans_pca.labels_)
purity_pca = purity_score(y_true, kmeans_pca.labels_)
rand_index_pca = adjusted_rand_score(y_true, kmeans_pca.labels_)

print(f'Sum Squared Error (PCA Data): {sse_pca}')
print(f'Purity (PCA Data): {purity_pca}')
print(f'Rand Index (PCA Data): {rand_index_pca}')
```

```python
#Q2.c

from sklearn.metrics import accuracy_score

# Function to assign labels to clusters based on the most frequent digit
def assign_labels_to_clusters(labels, clusters, num_clusters=10):
    assigned_labels = np.zeros_like(clusters)
    for i in range(num_clusters):
        cluster_mask = (clusters == i)
        cluster_labels = labels[cluster_mask]
        most_frequent_label = np.bincount(cluster_labels).argmax()
        assigned_labels[cluster_mask] = most_frequent_label
    return assigned_labels

# Assign labels to clusters for raw data
kmeans_raw_labels = assign_labels_to_clusters(y_train.flatten(), kmeans_raw.labels_)
accuracy_raw = accuracy_score(y_train.flatten(), kmeans_raw_labels)

# Assign labels to clusters for PCA-transformed data
kmeans_pca_labels = assign_labels_to_clusters(y_train.flatten(), kmeans_pca.labels_)
accuracy_pca = accuracy_score(y_train.flatten(), kmeans_pca_labels)

print(f'Accuracy of k-means on raw data: {accuracy_raw:.4f}')
print(f'Accuracy of k-means on PCA-transformed data: {accuracy_pca:.4f}')
```
